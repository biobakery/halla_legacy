```python
import strudel, halla
```

```python
s = strudel.Strudel( ) 
```

```python
X = s.randmat( shape = 100 ) 
Y = s.spike( X, strMethod = "parabola" ) 
aSpikeMethods = ["sine", "half_circle", "parabola", "cubic", "log", "vee", "linear"] 
```

```python
##Examples of spike types
import pylab
for strMethod in aSpikeMethods:
    pylab.figure( )
    strudel.scatter( X, s.spike(X, strMethod = strMethod ) ) 
```

```python
aAssociations = ["pearson", "spearman", "kw", "anova", "x2"]
aMIAssociations = ["norm_mi", "mi"] 
```

```python
for strMethod in aMIAssociations:
    print s.association( X, Y, strMethod = strMethod, bPval = True, bParam = False, bDiscretize = True ) ##You need discretization for mutual information based measures
```
